# 4. Variablen

* Variablen sind wie Behälter, in denen man Daten speichern kann.
* Eine Variable wird erstellt, wenn man ihr zum ersten Mal einen Wert gibt.
* In Python gibt es keine spezielle Anweisung (wie andere Programmiersprachen), um eine Variable zu erstellen.

```python
x = 7 # x is a variable of type int

y = "Hello, World!"  # y is a variable of type str

student = "H"  # student is a variable of type str

Student1 = 'Python'  # student1 is a variable of type str

Student2 = 15 # student2 is a variable of type int
```

## 4.2. Regeln für Python-Variablen:

* Ein Variablenname muss mit einem Buchstaben oder dem Unterstrich Zeichen beginnen.
* Ein Variablenname darf nicht mit einer Zahl beginnen.
* Ein Variablenname kann nur alphanumerische Zeichen und Unterstriche enthalten (A-z, 0-9 und _ ).
* Variablennamen unterscheiden zwischen Groß- und Kleinschreibung (age, Age und AGE sind drei verschiedene Variablen).
* Ein Variablenname darf keines der Python-Schlüsselwörter sein (`str`, `dict`, `def`, `set`, ...).

```python
# Legal variable names
# all variables are different and DO NOT override each other.
myvar = "python"
my_var = "python"
_my_var = "python"
myVar = "python"
MYVAR = "python"
myvar2 = "python"
```

```python
# Illegal variable names.
# Es wird ein Error angezeigt wenn du versuchst, den Code auszufuehren:
2myvar = "John"
my-var = "John"
my var = "John"
```

```python
# Multi Words Variable Names
myVariableName = "John" # Camel Case
MyVariableName = "John" # Pascal Case
my_variable_name = "John" # Snake Case
```

## 6.1. Integer (`int`)

Int, oder Ganzzahl, ist eine ganze Zahl, positiv oder negativ, ohne Dezimalstellen.

```python
x = 1
y = 35656222554887711
z = -3255522

print(type(x))
print(type(y))
print(type(z))
```

## 6.2. Float

Float oder "Gleitkommazahl" ist eine Zahl, positiv oder negativ, die eine oder mehrere Dezimalstellen enthält.

```python
x = 1.10
y = 1.0
z = -35.59

print(type(x))
print(type(y))
print(type(z))
```

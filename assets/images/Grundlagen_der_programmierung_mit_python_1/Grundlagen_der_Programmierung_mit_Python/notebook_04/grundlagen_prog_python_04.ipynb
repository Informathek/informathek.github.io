# 11. `if` und `else` Anweisung (statement)

## 11.1. `if` Anweisung

Logische Bedingungen aus der Mathematik:

* Gleich: a == b

* Nicht gleich: a != b

* Kleiner als: a < b

* Kleiner oder gleich: a <= b

* Größer als: a > b

* Größer oder gleich: a >= b



Diese Bedingungen können auf verschiedene Weisen verwendet werden, am häufigsten in "if-Anweisungen" und Schleifen.

Eine "if-Anweisung" wird durch Verwendung des Schlüsselworts "if" geschrieben.

```python
alter = int(input("Was ist dein Alter? "))

if alter < 18:
  print("Du bist minderjährig")
```

## 11.2. Indentation


```python
# Falsches Syntax: wird ein error angezeigt.
if alter < 18:
print("Du bist minderjährig")
```

```python
# Mehere Anweisungen
if alter < 18:
  print("Du bist minderjährig.")
  print("Dein Zugang ist beschränkt.")
```

```python
alter = int(input("Was ist dein Alter? "))

if alter < 18:
  print("Du bist minderjährig.")
print("Vielen Dank für die Überprüfung deines Alters!")
```

## 11.3. `else` Anweisung


```python
# Alter des Benutzers abfragen
alter = int(input("Geben Sie Ihr Alter ein: "))


# Ueberpruefen, ob die Person minderjaehrig ist
if alter < 18:
    print("Du bist minderjährig. Zugang ist beschränkt.")
else:
    print("Du bist volljährig. Willkommen!")


# Diese Nachricht wird immer angezeigt
print("Vielen Dank für die Überprüfung deines Alters!")
```

```python
# if-else mit logischen Operatoren

# Benutzer nach ihrem Alter und ihrer Staatsbürgerschaft fragen
alter = int(input("Geb Dein Alter ein: "))
Nationalitaet = input("Gib mir deine Nationalitaet: ")

# Ueberpruefen, ob die Person wahlberechtigt ist
if alter >= 18 and Nationalitaet == "Deutsch":
    print("Du bist wahlberechtigt.")

# Ueberprüfen, ob das Alter unter 18 liegt oder keine Staatsbuergerschaft besteht
else:
    print("Du bist nicht wahlberechtigt.")
```

## 11.4. elif

```python
# Verschachtelte else Bedingung

zahl = int(input("Geben Sie eine Zahl ein: "))

if zahl > 0:
    print("Die Zahl ist positiv.")
else:
    if zahl < 0:
        print("Die Zahl ist negativ.")
    else:
        print("Die Zahl ist null.")
```

```python
# Besser ist, wenn wir elif bunutzen

zahl = int(input("Geben Sie eine Zahl ein: "))

if zahl > 0:
    print("Die Zahl ist positiv.")
elif zahl < 0:
    print("Die Zahl ist negativ.")
else:
    print("Die Zahl ist null.")
```

# 12. `while` Schleife (loop)

Python hat zwei grundlegende Schleifen Befehle:

* `while`-Schleifen: Heute
* `for`-Schleifen: Spaeter

```python
print("Hier wird eine Nachricht 5 Mal wiederholt.")

zaehler = 1

while zaehler <= 5:
    print("Hallo Schüler nummer ", zaehler)
    zaehler = zaehler + 1


# Verabschiedung
print("Das war die Motivation mit While-Schleifen.")
```

## 12.1. Break

```python
print("Willkommen zum break-Beispiel!")

# Hauptschleife
while True:
    eingabe = input("Gib 'q' ein, um die Schleife zu beenden, oder etwas anderes zum Fortfahren: ")

    # Ueberprüfen, ob die Eingabe 'q' ist
    if eingabe == 'q':
        print("Schleife wird beendet.")
        break  # Die Schleife wird hier beendet

    else:
        print("Du hast etwas anderes eingegeben. Die Schleife wird fortgesetzt.")

# Verabschiedung
print("Das war das break-Beispiel. Auf Wiedersehen!")
```

```python
# Task: Ratspiel Simulation
import random

# Zufällige Zahl zwischen 1 und 10 generieren
zielzahl = random.randint(1, 10)
```

```python
# Begrussungsnachricht
print("Willkommen zum Zahlenraten!")

# Initialisierung der Rateversuche
versuche = 0

# Hauptschleife des Spiels
while True:
    # Benutzer zur Eingabe einer Zahl auffordern
    geratene_zahl = int(input("Gib eine Zahl zwischen 1 und 10 ein: "))

    # Versuche erhoehen
    versuche = versuche + 1

    # Ueberpruefen, ob die geratene Zahl mit der Zielzahl übereinstimmt.
    if geratene_zahl == zielzahl:
        print(f"Herzlichen Glückwunsch! Du hast die Zahl {zielzahl} in {versuche} Versuchen erraten.")
        break
    elif geratene_zahl < zielzahl:
        print("Die gesuchte Zahl ist größer.")
    else:
        print("Die gesuchte Zahl ist kleiner.")

# Das Spiel ist beendet
```

## 3.2. Comments

* explains Python code.
* makes the code more readable.
* prevents execution when testing code.
* Comments starts with a **#**, and Python will ignore them:

```python
# This is a comment
print("Hello, World!")

print("Hello, World!") # This is a comment
```

```python
# This is a comment
# written in
# more than just one line
print("Hello, World!")
```

```python
# Using a multi-line comment:
"""
This is a comment
written in
more than just one line
"""
print("Hello, World!")
```

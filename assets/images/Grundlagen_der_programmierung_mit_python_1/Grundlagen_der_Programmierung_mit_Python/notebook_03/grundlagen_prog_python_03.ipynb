# 07. Booleans

Booleans sind Datentypen und repräsentieren einen von zwei Werten: Wahr (`True`) oder Falsch (`False`).

* In der Programmierung wollen wir oft wissen, ob ein Ausdruck wahr oder falsch ist.

* Wir können jeden Ausdruck in Python auswerten und eine von zwei Antworten erhalten: Wahr oder Falsch.

* Wenn wir zwei Werte vergleichen, wird der Ausdruck ausgewertet und Python liefert die boolesche Antwort

```python
vater = 40  # Alter in Jahre
kind = 10 # Alter Jahre

groesser = (vater > kind)
kleiner = (vater < kind)
gleich = (vater == kind)

print(groesser)
print(kleiner)
print(gleich)
```

```python
print(type(groesser))
```

```python
print(vater >= kind)
print(vater < kind)
print(vater == kind)
```

```python
x = 5
y = 5

print(x >= y)
```

```python
print(x == y)
```

```python
print(x != y)
```

```python
print(5 == 5)
print((10 * 2) != 21)
```

# 8. Operators

Operatoren werden verwendet, um Operationen an Variablen und Werten durchzuführen.

```python
x = 10 + 5

print(x)
```

Python unterteilt die Operatoren in folgende Gruppen:

* Arithmetische Operatoren (+, -, *, /)
* Zuweisungsoperatoren (=)
* Vergleichsoperatoren (<, >, ==, !=)
* Logische Operatoren (and, or, not)
* Identitätsoperatoren (is, is not)
* Mitgliedschaftsoperatoren (in, not in)
* Bitweise Operatoren (&, |, ^, ~)

### 8.1. Logische Operatoren

```python
x = 5 > 2
print(x)
```

```python
ramez = 18
reem = 15
kind = 10

ramez_ist_der_alterste = (ramez > reem) and (ramez > kind)

print(ramez_ist_der_alterste)
```

```python
mathe = 10
physik = 3
chemie = -5

# normale Schule: ein model bstanden, dann schule bestanden
normale_schule_bestanden = (mathe > 0) or (physik > 0) or (chemie > 0)
print(normale_schule_bestanden)

# schwere Schule: ein modul durchgefallen, dann  schule durchgefallen
schwere_schule_bestanden = (mathe > 0) and (physik > 0) and (chemie > 0)
print(schwere_schule_bestanden)
```

# 9. Lists

```python
obst_liste = ["Anananas", "Erdberen", "Mango", 42]

print(obst_liste)
print(type(obst_liste))
```

## 9.1. Definition
* Listen werden verwendet, um mehrere Elemente in einer einzigen Variable zu speichern.

* Listen sind eine von 4 integrierten Datentypen in Python, die verwendet werden, um Sammlungen von Daten zu speichern. Die anderen 3 sind `Tuple`, `Set` und `Dictionary`, jeweils mit unterschiedlichen Eigenschaften und Verwendungszwecken.

* Listen werden mit eckigen Klammern erstellt: `[...]`

## 9.2. Eigenschaft: geordert

Wenn wir sagen, dass Listen geordnet sind, bedeutet dies, dass die Elemente eine festgelegte Reihenfolge haben, und diese Reihenfolge wird sich nicht ändern.

Wenn Sie neue Elemente zu einer Liste hinzufügen, werden die neuen Elemente am Ende der Liste platziert.

```python
print(obst_liste)
```

## 9.3 Eigenschaft: Indexing

Die Indexierung in Python beginnt bei 0, was bedeutet, dass das erste Element in einer Sequenz den Index 0 hat, das zweite Element den Index 1 usw.

```python
print(obst_liste[0])

print(obst_liste[3])
```

## 9.4. Eigenschaft: veraenderbar

Eine Liste ist veränderbar, bedeutet, dass wir Elemente in der Liste ändern, hinzufügen und entfernen können, nachdem sie erstellt wurde.

```python
print(obst_liste)
```

```python
obst_liste[0] = "Ananas"
```

```python
print(obst_liste)
```

```python
obst_liste[3] = 40
```

```python
obst_liste
```
